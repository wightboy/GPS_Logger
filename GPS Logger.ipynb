```python
import pandas as pd
import folium
from datetime import datetime
```

```python
df = pd.read_csv("C:\\Users\\Nicholas\\20160929.txt")
df['str time'] = df['time']
def func(row):
    row = row.replace("T", " ")
    row = row.replace("Z", " ")
    row = row.replace("-", " ")
    row = row.replace(":", " ")
    row = row.replace(" ", "")
    return row

df['time'] = df['time'].apply(func)
df['time'] = pd.to_datetime(df['time'], format='%Y%m%d%H%M%S')
df.head()
```

```python
m = folium.Map(location=[df['lat'][0], df['lon'][0]], zoom_start=13)
coordinates = []
rgb = 200 / float(len(df))
for x in range(0, len(df)-1):
    n = (rgb * x) / 2
    coordinates = []
    coordinates.append([df['lat'][x], df['lon'][x]])
    coordinates.append([df['lat'][x+1], df['lon'][x+1]])
    my_PolyLine=folium.PolyLine(locations=coordinates,weight=5,
                                color='rgb('+str(int(n))+', '+str(int(n))+', '+str(int(n))+')')
    m.add_children(my_PolyLine)
    folium.Marker([df['lat'][x], df['lon'][x]], 
                  popup=str(df['time'][x].time())+ "\n" + str(df['accuracy'][x])+"m").add_to(m)
    folium.CircleMarker(location=[df['lat'][x], df['lon'][x]], radius=df['accuracy'][x],
                     color='#3186cc', fill_color='#3186cc').add_to(m)
    
folium.Marker([df['lat'][len(df)-1], df['lon'][len(df)-1]], 
              popup=(str(df['time'][len(df)-1].time())+ "\n" + str(df['accuracy'][len(df)-1])+"m")).add_to(m)
folium.CircleMarker(location=[df['lat'][len(df)-1], df['lon'][len(df)-1]], radius=df['accuracy'][len(df)-1],
                     color='#3186cc', fill_color='#3186cc').add_to(m)

m.save('line_example.html')
```
